```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
fuel_types = ['Gasoline', 'Ethanol (E-85)','Diesel Fuel',  'Electricity', 'Compressed Natural Gas (CNG)']
number_of_vehicles = [351304, 4645,3505,  20983, 678]
df = pd.DataFrame({'Fuel': fuel_types, 'Number of Vehicles': number_of_vehicles})

# Create a figure and axes
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the pie chart with 15-degree rotation
wedges, texts, autotexts = ax.pie(df['Number of Vehicles'], labels=df['Fuel'], autopct='%1.1f%%', startangle=15, colors=sns.color_palette("pastel"))

# Add a circle at the center to create the donut shape
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Add a title
ax.set_title('Vehicle Distribution by Fuel Type')

# Show the pie chart
plt.show()
```
